KDE (Kernel Density Estimate) // (Estimativa de densidade do kernel)

É um grafico usado para Visualizar a densidade de probabilidade de uma variavel númerica.

Em Estatística, estimativa de densidade por Kernel (EDK) é uma forma não-paramétrica para estimar a Função densidade de probabilidade (FDP) de uma variável aleatória. Estimativa da densidade por Kernel é um problema fundamental de suavização de dados onde inferências sobre a população são feitas com base em uma amostra de dados finita.

Pré-processamento e bases de treinamento/teste


ATENÇÃO!!!
Quando a base de dados tem colunas com atributos diferentes mas equivalentes
(Exemplo: Masculino e Feminino, São variaveis diferentes, mas equivalentes(nenhum é melhor que o outro)).
Nesses casos é preciso Criar novas colunos para cada atributo(para prepara os dados, (no uso de Machine Learning os dados tem que ser Númericos/quantitativos))
Exemplo da coluna "Genero" que possui os valor Male e Female, Deve ser Criado as colunas:
E vai receber como atributos 0 ou 1. 1 => se caso for "sim", e 0 => case "não"

Male | Female
1    |   0
0    |   1

Para fazer esse pocesso pode ser usado a biblioteca sklearn.preprocessing impontando OnheHotEncoder

```python
# Importando Biblioteca
from sklearn.preprocessing import OneHotEncoder
```

```python
# Primero, criar um dataframe com as colunas categoricas a seram modificadas
X_cat = employee_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
X_cat
```

```python
# Segundo, criar um Objeto da classe OneHotEncoder
onehotencoder = OneHotEncoder()
```

```python
# Terceito, Tranformar os dados categoricos em Númericos, Coma criação de novas colunas.
# fit_transform(dataframe) : essa função faz a transformação dos dados do dataframe indicado nos parametros
# .toarray() : faz o converssão o rasultado em um array.
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
X_cat.shape
```

Antes da tranformação as Linhas x Colunas => 1470 rows × 6 columns
Depois da tranformação as Linhas x Colunas => 1470 rows x 26 columns

```python
# Transforma o array em um data frame
X_cat = pd.DataFrame(X_cat)
type(X_cat)
```

```python
# Verificar novas colunas e valores
X_cat.head()
```

```python
# varificar os valores unicos da coluna(valores que aparecem na coluna BusinessTravel)
employee_df['BusinessTravel'].unique()
```

```python
# Criar um data frame só com variaveis numericas:
X_numerical = employee_df[['Age', 'DailyRate', 'DistanceFromHome',	'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',	'JobLevel',	'JobSatisfaction',	'MonthlyIncome',	'MonthlyRate',	'NumCompaniesWorked',	'OverTime',	'PercentSalaryHike', 'PerformanceRating',	'RelationshipSatisfaction',	'StockOptionLevel',	'TotalWorkingYears'	,'TrainingTimesLastYear'	, 'WorkLifeBalance',	'YearsAtCompany'	,'YearsInCurrentRole', 'YearsSinceLastPromotion',	'YearsWithCurrManager']]
X_numerical
```

```python
# A Função pd.concat([]) recece dois data frame do Pandas e faz a junção/concatena deles.
X_all = pd.concat([X_cat, X_numerical], axis = 1)
X_all
```

É preciso fazer uma Narmalização dos dados, Pois existe colunas com valores Muito Altos, e com isso, o algoritome de Machine Learning pode priorizar esses valores no Treinamento. E isso pode influenciar os resultado de forma negativa. 
Para isso é precisor normalizar os dados e deixar eles em uma "mesma escala".

```python
# Importa a Biblioteca sklearn. preprocessing import MinMaxScaler
# MinMaxScaler : vai analisar os valores minimos e maximos e vai colocar os valores na mesma escala.
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# O objeto scaler.fit_transform(X_all) : vai analisar o data frame X_all(quem possui todos o dados(Númericos)) e vai tranforma para uma mesma escala
X = scaler.fit_transform(X_all)
```

```python
# Exibição do valores depois da normalização(escala)
# Esses valores seram os previsores para o Treinamento de maquina
X
```

```python
# Attrition vai ser os valor de classe
y = employee_df['Attrition']
y
```

```python
# importa a biblioteca sklearn.model_selection import train_test_split , para dividir o data frame em dados de traine e teste
from sklearn.model_selection import train_test_split
X_train, X_test , y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
# Verificar a composição das bases criada para Treino
X_train.shape, y_train.shape
```

```python
# Verificar a composição das bases criada para Teste
X_test.shape, y_test.shape
```

```python
# Contagem da frequencia de atributos de uma coluna. Veirificar o Balanceamento de classe.
employee_df.groupby(["Attrition"]).count()
```

Redes Neurais Artificiais

```python
import tensorflow as tf
```

```python
X_train.shape
```

(units=) esse parâmetro recebe o nomero de neurorios da camada.
A primeira camanda recebe a quantidade de entradas diferentes da variavel de "X_Treino"
Exemplo: X_train.shape : (1102, 50), "units" pode colocar 50 ou (50+1)/2 => a soma da quantidade de entrada e saida divido por 2. 

```python
# Comando para criar uma seguencia de camadas, para o uso de Rede reural.
rede_neural = tf.keras.models.Sequential()
```

```python
# Criando camanda de entrada Unitis vai receber (50+1)/2 =>25
# Parâmetro input_shape recebe a quantidade de entradas: X_train.shape : (1102, 50)
rede_neural.add(tf.keras.layers.Dense(units=25, activation='relu', input_shape=(50,)))
rede_neural.add(tf.keras.layers.Dense(units=25, activation='relu'))
rede_neural.add(tf.keras.layers.Dense(units=25, activation='relu'))
rede_neural.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
# Visualizar composição das camandas da rede neural
rede_neural.summary()
```

```python
# Criar um compilador 
# optimizer='Adam' => É um parâmetro para Melhorar a compulação, e "Adam" é o mais usado/comum 
# loss='binary_crossentropy' => Como deve ser calculada as perdas . como o resultado deve ser 0 ou 1, é usado binary_crossentropy 
# metrics=['accuracy'] => Parâmentro de medida de acertos. Em Porcentagem.
rede_neural.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Treinar a rede neural com os valores de X_train e y_train(valores gerado em etapas anteriores)
# Parâmetro "epochs=" recebe a quantidade epocas/vezes que sera refeito o teste para ajustar os pesos.
rede_neural.fit(X_train, y_train, epochs=200)
```

```python
# Criar as precisões com base nos resultados da rede neural
y_pred = rede_neural.predict(X_test)
y_pred
```

```python
y_pred = (y_pred >= 0.5)
y_pred
```

```python
cm = confusion_matrix(y_pred, y_test)
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
# Verficando a % de acertos, usando classification_report
print(classification_report(y_test, y_pred))
```
